```python
nfl_data = pd.read_csv("NFL Play by Play 2009-2016 (v3).csv")
nfl_data.head()
```

```python
missing_values = nfl_data.isnull().sum()
missing_values[0:10]
```

```python
nfl_data.dropna()
```

```python
columns_with_na_dropped = nfl_data.dropna(axis = 1)
columns_with_na_dropped.head()
```

```python
# Just how much data did we lose?
print("Colums in original dataset: %d \n" % nfl_data.shape[1])
print("Columns with na's is dropped: %d" % columns_with_na_dropped.shape[1])
```

```python
# get a small subset of the NFL dataset
subset_nfl_data = nfl_data.loc[:, 'EPA': 'Season'].head()
subset_nfl_data
```

```python
# replace all NA's the value that comes directly after it in the same column
# then replace all teh remaining na's with )'s
subset_nfl_data.fillna(method = 'bfill', axis = 0).fillna(0)
```

```python
# import libraries

# Take a first look at the data

# How many missing data points do we have?

# figure out why data is missing

# drop missing values : rows

# drop missing values : columns

# fill in missing values automatically
```

```python
missing_values = nfl_data.isnull().sum()

total_number = np.product(nfl_data.shape)
total_missing = missing_values.sum()

print(total_number)
print(total_missing)
print('Percent: ', (total_missing/total_number)*100)
```
