```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('titanic_train.csv')
train
```

```python
train.describe()
```

```python
train.info()
```

```python
train.Survived.value_counts()
```

```python
train[['Pclass', 'Survived']].groupby('Pclass').mean().Survived.plot(kind = 'bar', title = 'TBD')
plt.xlabel('Pclass')
plt.ylabel('Survival Probability')
```

```python
survived = train[['Survived', 'Pclass']].groupby('Pclass').sum()
Probability = survived['Survived']/survived['Survived'].sum()
Probability
```

```python
train[['Survived', 'Pclass']].groupby('Pclass').mean().Survived.plot(kind = 'pie')
```
